```python
import json
import numpy as np
import pandas as pd
from itertools import product
```

```python
pd.read_csv("data/2022/input/skf/embeddings/darknet_features_20221021.csv").head()
```

```python
strats = [["features", "darknet"]]
```

```python
jp = "data/2022/input/skf/stratification/stratification.json"
with open(jp, 'r') as fd:
    splits = json.load(fd)
```

```python
len(splits["20221021"]["darknet"]), len(splits["20221021"]["darknet"][0])
```

```python
days = list(splits.keys())
days.sort()
days
```

```python
iters = product(days, strats, np.arange(10))
for day, (strat, label), fold in iters:
    file_path = f"data/2022/input/skf/embeddings/darknet_{strat}_{day}.csv"
    df = pd.read_csv(file_path)
    train_ips = splits[day]["darknet"][0][0]
    test_ips = splits[day]["darknet"][0][1]
    df_train = df[df.src_ip.isin(train_ips)]
    df_test = df[df.src_ip.isin(test_ips)]
    df_train.to_csv(
        f"data/2022/input/embeddings/out/k3/train/{label}_{day}_fold0{fold}.csv"
    )
    df_test.to_csv(
        f"data/2022/input/embeddings/out/k3/test/{label}_{day}_fold0{fold}.csv"
    )
```
